```python
df2['Average_change_in_10years_per_country'] = df2[["F2013","F2014","F2015","F2016","F2017","F2018","F2019","F2020","F2021","F2022"]].mean(axis=1)
```

# Average change in 10 years where average change>1.5

```python
Average_change_in_10years= df2['Average_change_in_10years_per_country']
```

```python
Average_change_in_10years[Average_change_in_10years>1.5].plot(kind='bar',x=df2['Country'])
```

```python
df3 = df2[['Country','Average_change_in_10years_per_country']]
```

```python
plt.figure(figsize=(40,40))
sns.barplot(x=df2['Average_change_in_10years_per_country']>1.5,y=df2['Country'],palette = 'viridis')
```

# Max and Min temperature increase in the last 10 years along with the countries

```python
max_temp = df3['Average_change_in_10years_per_country'].max()
```

```python
max_temp
```

```python
 max_temp = df3.query('Average_change_in_10years_per_country==Average_change_in_10years_per_country.max()')
```

```python
 min_temp = df3.query('Average_change_in_10years_per_country==Average_change_in_10years_per_country.min()')
```

```python
max_temp
```

```python
min_temp
```

```python
min_max_temps = pd.concat([max_temp,min_temp],axis=0)
```

```python
min_max_temps
```

```python
sns.barplot(x=min_max_temps['Country'],y =min_max_temps['Average_change_in_10years_per_country'])
```

```python
top_5_countries_with_highest_temp = df3.nlargest(5,['Average_change_in_10years_per_country'])
```

```python
sns.barplot(x= top_5_countries_with_highest_temp['Country'],y = top_5_countries_with_highest_temp['Average_change_in_10years_per_country'])
```

```python
top_5_countries_with_highest_temp
```

```python
top_5_countries_with_lowest_temp_rise = df3.nsmallest(5,['Average_change_in_10years_per_country'])
```

```python
top_5_countries_with_lowest_temp_rise
```

```python
sns.barplot(x= top_5_countries_with_lowest_temp_rise['Country'],y = top_5_countries_with_lowest_temp_rise['Average_change_in_10years_per_country'])
```
